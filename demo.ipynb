```python
import json
pairs = []
with open("../nfd/train_deepl.jsonl") as train, open("../nfd/dev_deepl.jsonl") as dev, open("../nfd/test_deepl.jsonl") as test:
    for line in train.readlines()+test.readlines()+dev.readlines():
        l = json.loads(line)
        for ev in l["evidence"]:
            pairs.append((ev,l))
```

```python
len(pairs)
```

```python
neccessary_wiki = set()
for ev,l in pairs:
    for e in ev:
        neccessary_wiki.add(e[2])
```

```python
neccessary_wiki
```

```python
wikimap = {}
```

```python
cswiki = {}
with open("../cswiki.jsonl") as cswiki:
    for line in cswiki:
        line = json.loads(line)
        if line["id"] not in neccessary_wiki:
            continue
        wikimap[line["id"]]=line["text"]
```

```python
wikimap
```

```python
duplicates = set()
with open("feverdata.jsonl", "w") as feverdata:
    for ev, l in pairs:
        try:
            out = {"fever_id": l["id"],"claim": l["claim_en"], "claim_cs": l["claim"], "label": l["label"], "evidence_cs": {e[2]: wikimap[e[2]] for e in ev}, "evidence": {e[4]: e[4] for e in ev}}
            duplicate = tuple(out["claim"] + out["evidence_cs"])
            if duplicate in duplicates:
                continue
            duplicates.add(duplicate)
            print(json.dumps(out, ensure_ascii=False),file=feverdata)
        except KeyError:
            pass
```
